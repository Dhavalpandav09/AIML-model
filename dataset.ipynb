```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("Employee_HR.csv")

X = data[['time_spent_company']]
y= data['Salary_INR']

model = LinearRegression()
model.fit(X,y)
predicted_scores = model.predict(X)

mae = mean_absolute_error(y,predicted_scores)
mse = mean_squared_error(y,predicted_scores)
rmse = np.sqrt(mse)
r2 = r2_score(y,predicted_scores)

print("Mean Absolute Error",round (mae,2))
print("Mean Squared Error",round (mse,2))
print("Root Mean Squared Error",round(rmse,2))
print("R^2 (model accurcy ):",round(r2,2))


plt.figure(figsize=(10,6)) #10 is width , and 6 is height
plt.hist(data["time_spent_company"],bins=30, color='skyblue',edgecolor='black')
plt.title("salary of employee")
plt.xlabel("salary")
plt.ylabel("emp")
plt.grid(True)
plt.show()


#scatter 

plt.figure(figsize=(10,6)) #10 is width , and 6 is height
plt.scatter(X,y,color='skyblue',label='salary')
plt.plot(X,predicted_scores,color="red",label="preadicted scores(line)")
plt.title("model prection vs actual scores")
plt.xlabel("time speneded")
plt.ylabel("final output")
plt.grid(True)
plt.show()


new_time = 9
predicted_new_score = model.predict([[new_time]])
print(f"predicted final salary for {new_time} houes is {predicted_new_score}")
```
